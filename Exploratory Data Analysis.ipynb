```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("D:\ecommerce_analysis\Ecommerce-Analysis-With-EDA-and-MySQL\Revenue Trends.csv")
df1.head()
```

# Monthly Revenue Growth

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x='month', y='revenue', marker='o', color='royalblue')
plt.title('Monthly Revenue Growth', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Revenue vs Order Volume

```python
fig, ax1 = plt.subplots(figsize=(12,6))

color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue (R$)', color=color)
ax1.plot(df1['month'], df1['revenue'], color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Orders Count', color=color)
ax2.plot(df1['month'], df1['orders_count'], color=color, marker='x')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Revenue vs Order Volume', fontsize=16)
plt.xticks(rotation=45)
fig.tight_layout()
plt.show()
```

```python
df2 = pd.read_csv("D:\ecommerce_analysis\Ecommerce-Analysis-With-EDA-and-MySQL\Top Sell Item.csv")
df2.head()
```

# Product Performance: Revenue vs Units Sold

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df2, x='items_sold', y='total_revenue', 
                size='total_revenue', hue='total_revenue',
                sizes=(50, 500), palette='viridis')
plt.title('Product Performance: Revenue vs Units Sold', fontsize=16)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Total Revenue (R$)', fontsize=12)
plt.legend(title='Revenue', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

# Top 5 Products by Revenue

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,6))

# Plot 1: Top products by revenue
sns.barplot(data=df2.head(5), x='total_revenue', y='product_id', ax=ax1, palette='Blues_d')
ax1.set_title('Top 5 Products by Revenue')
ax1.set_xlabel('Total Revenue (R$)')

# Plot 2: Revenue distribution
sns.boxplot(data=df2, x='total_revenue', ax=ax2, color='lightblue')
ax2.set_title('Revenue Distribution Across Products')
ax2.set_xlabel('Revenue (R$)')

plt.tight_layout()
plt.show()
```
